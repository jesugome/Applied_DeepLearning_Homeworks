## 4.1 Little Red Riding Hood Network

Train a neural network to solve the  Little Red Riding Hood problem in sklern and Keras. Try the neural networ with different inputs and report the results.

<img src='images/littlered.jpg'>

###### Solving with keras.

````python
import tensorflow as tf
````

````python
import numpy as np
import pylab as pl
import keras

%matplotlib inline


# Create training data.
x_train = np.array([[1,1,0,0],[0,1,1,0],[0,0,0,1]])
y_train = np.array([[1,0,0,0],[0,0,1,1],[0,1,1,0]])

# Neural net structure. 
from keras.models import Sequential
from keras.layers import Dense, Activation

neuralNetKeras = Sequential()

neuralNetKeras.add(Dense(units=3, input_dim=4))
neuralNetKeras.add(Activation('sigmoid'))

neuralNetKeras.add(Dense(units=4))
neuralNetKeras.add(Activation('sigmoid'))

print('The neural net constructor is as follows: ')
neuralNetKeras.summary()
````

````python
from keras.optimizers import SGD

# Compile the structure. 
neuralNetKeras.compile(loss = 'binary_crossentropy',
              optimizer = SGD(lr=0.01, momentum=0.9, decay=0.0, nesterov=False),
              metrics = ['accuracy'])

# Traning the neural net. 
neuralNetKeras.fit(x_train, y_train, epochs=5000, verbose=2)
````

````python
# Test a sample output.
print('-------------------------------------')
print('The model outputs are: ')
print(np.round(neuralNetKeras.predict(np.array([[1,1,0,0]])),4))
print('-------------------------------------')
print('After round the outputs are: ')
print(np.round(neuralNetKeras.predict(np.array([[1,1,0,0]])),2))
print('-------------------------------------')
print('Right output is: ')
print(np.array([[1,0,0,0]]))
````

Solving with Keras the output is very close to the right output. The result is inaccurate due to lack of samples. Nevertheless testing over a sample in the train the result is equal to the expected response

###### Solving with sklern.

````python
import numpy as np
import pylab as pl
import keras

%matplotlib inline
# Create training data.
x_train = np.array([[1,1,0,0],[0,1,1,0],[0,0,0,1]])
y_train = np.array([[1,0,0,0],[0,0,1,1],[0,1,1,0]])
````

````python


##### Multiclass Classificator

# Neural net structure. 
from sklearn.neural_network import MLPClassifier
neuralNetC = MLPClassifier(solver='lbfgs', alpha=1e-5,
                   hidden_layer_sizes=(3), random_state=1, activation='logistic')

neuralNetC.fit(x_train,y_train)
````

````python
# Test a sample output.
print('-------------------------------------')
print('The model outputs are: ')
print(neuralNetC.predict(np.array([[1,1,0,0]])))
print('-------------------------------------')
````

````python
# Multiclass regression Problem( Index close to 0 and 1)


# Neural net structure. 
from sklearn.neural_network import MLPRegressor
neuralNetR = MLPRegressor(solver='lbfgs', alpha=1e-5,
                   hidden_layer_sizes=(3), random_state=1, activation='logistic')

neuralNetR.fit(x_train,y_train)
````

````python
# Test a sample output.
print('-------------------------------------')
print('The model outputs are: ')
print(np.round(neuralNetR.predict(np.array([[1,1,0,0]])),4))
print('-------------------------------------')
print('After round the outputs are: ')
print(np.round(neuralNetR.predict(np.array([[1,1,0,0]])),2))
print('-------------------------------------')
````

Solving with Sklearn the output is very close to the right output too. The result is inaccurate due to lack of samples. Nevertheless testing over a sample in the train the result is equal to the expected response

## 4.2 Boston House Price Prediction

In the next questions we are going to work using the dataset *Boston*. This dataset measures the influence of socioeconomical factors on the price of several estates of the city of Boston. This dataset has 506 instances, each one characterized by 13 features:

* CRIM - per capita crime rate by town
* ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
* INDUS - proportion of non-retail business acres per town.
* CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
* NOX - nitric oxides concentration (parts per 10 million)
* RM - average number of rooms per dwelling
* AGE - proportion of owner-occupied units built prior to 1940
* DIS - weighted distances to five Boston employment centres
* RAD - index of accessibility to radial highways
* TAX - full-value property-tax rate per 10,000 USD
* PTRATIO - pupil-teacher ratio by town
* B - $1000(Bk - 0.63)^2$ where $Bk$ is the proportion of blacks by town
* LSTAT - % lower status of the population

Output variable:
* MEDV - Median value of owner-occupied homes in 1000's USD

**Note:** In this exercise we are going to predict the price of each estate, which is represented in the `MEDV` variable. It is important to remember that we are always aiming to predict `MEDV`, no matter which explanatory variables we are using. That means, in some cases we will use a subset of the 13 previously mentioned variables, while in other cases we will use all the 13 variables. But in no case we will change the dependent variable $y$.

````python
## Adding the target variable. 

boston['MEDV'] = bostonRaw.target

## Descrbing boston dataset.
round(boston.describe(),2)
````

## 4.4 Modeling linear and non linear relationships

* Generate two new subsets filtering these characteristics:
     * $D_1$:  $X = \textit{'RM'}$, $y = \textit{'MEDV'}$
     * $D_2$:  $X = \textit{'LSTAT'}$, $y = \textit{'MEDV'}$
* For each subset, generate a training partition and a test partition using a ratio of $ 70 \% - 30 \% $
* Train a linear regression model on both subsets of data:
     * Report the mean square error on the test set
     * Print the values of $ w $ and $ w_0 $ of the regression equation
     * Generate a graph where you visualize the line obtained by the regression model in conjunction with the training data and the test data
* How does the model perform on $ D_1 $ and $ D_2 $? Why?


````python
### Creating subsets.
x1 = boston['RM']
y1 = boston['MEDV']

x2 = boston['LSTAT']
y2 = boston['MEDV']
````

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
````

````python
##### Analysis for model 1.

x1train,x1test,y1train,y1test = train_test_split(x1,y1,test_size=0.30,random_state=22)

# Polinomial fit to a line.
linearModel=LinearRegression()

linearFit= linearModel.fit(x1train.values.reshape(-1,1),y1train.values.reshape(-1,1))

# Linear fit plot. 
plt.figure(figsize=(15,15))
plt.plot(x1train, y1train, 'bo',label="Train")
plt.plot(x1test, y1test, 'bo',color='orange',label="Test")
x=x1train
b0=linearFit.intercept_
b1=linearFit.coef_[0]
plt.plot(x,b0+b1*x,color="red",label="Linear")
plt.xlabel(x1.name)
plt.ylabel(y1.name)
plt.legend()
plt.title('Model 1 Fitting')

# Performance measuring(MSE).

y1pred = linearModel.predict(x1test.values.reshape(-1,1))
MSE= metrics.mean_squared_error(y1pred,y1test)

print('------------------------------')
print('The equation of linear regression has: \n ')

print('w0 is',b0)
print('w is',b1)

print('------------------------------')

print('The MSE of the model is: ',MSE)
````

````python

##### Analysis for model 2.

x2train,x2test,y2train,y2test = train_test_split(x2,y2,test_size=0.30,random_state=22)

# Polinomial fit to a line.
linearModel=LinearRegression()

linearFit= linearModel.fit(x2train.values.reshape(-1,1),y2train.values.reshape(-1,1))

# Linear fit plot. 
plt.figure(figsize=(15,15))
plt.plot(x2train, y1train, 'bo',label="Train")
plt.plot(x2test, y1test, 'bo',color='orange',label="Test")
x=x2train
b0=linearFit.intercept_
b1=linearFit.coef_[0]
plt.plot(x,b0+b1*x,color="red",label="Linear")
plt.xlabel(x2.name)
plt.ylabel(y2.name)
plt.legend()
plt.title('Model 2 Fitting')

# Performance measuring(MSE).

y2pred = linearModel.predict(x2test.values.reshape(-1,1))
MSE= metrics.mean_squared_error(y2pred,y2test)

print('------------------------------')
print('The equation of linear regression has: \n ')

print('w0 is',b0)
print('w is',b1)

print('------------------------------')

print('The MSE of the model is: ',MSE)
````

````python
### Partitioning all data.
xtrain,xtest,ytrain,ytest = train_test_split(boston.drop('MEDV',axis=1),boston['MEDV'],test_size=0.30,random_state=22)
````

````python
### Linear Regression. 

# Polinomial fit to a line.
linearModel=LinearRegression()

linearFit= linearModel.fit(xtrain,ytrain)

ypred = linearModel.predict(xtest)
MSE= metrics.mean_squared_error(ypred,ytest)

print('------------------------------------------')

print('The MSE of linear fit regression is: ',MSE)

print('------------------------------------------')
````

````python
### MLP Regressor. 

from sklearn.neural_network import MLPRegressor

mlp=MLPRegressor(hidden_layer_sizes=(128,512), solver='adam' , max_iter=1000)

mlpFit = mlp.fit(xtrain,ytrain)

ypred = mlpFit.predict(xtest)

MSE= metrics.mean_squared_error(ypred,ytest)

print('------------------------------------------')

print('The MSE of linear MultiLayerPerceptron is: ',MSE)

print('------------------------------------------')
````

````python
print('------------------------------------------')
print('------------------------------------------')

print('           Now scaling data')

print('------------------------------------------')
print('------------------------------------------')
````

````python
### Scaling data. 

from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
sc_x.fit(boston.drop('MEDV',axis=1))

xtrainNor = sc_x.transform(xtrain)
xtestNor = sc_x.transform(xtest)
````

````python
### Linear Regression. 

# Polinomial fit to a line.
linearModel=LinearRegression()

linearFit= linearModel.fit(xtrainNor,ytrain)

ypred = linearModel.predict(xtestNor)
MSE= metrics.mean_squared_error(ypred,ytest)

print('------------------------------------------')

print('The MSE of NEW linear fit regression is: ',MSE)

print('------------------------------------------')
````

````python
### MLP Regressor. 

from sklearn.neural_network import MLPRegressor

mlp=MLPRegressor(hidden_layer_sizes=(128,512), solver='adam' , max_iter=1000)

mlpFit = mlp.fit(xtrainNor,ytrain)

ypred = mlpFit.predict(xtestNor)
MSE= metrics.mean_squared_error(ypred,ytest)

print('------------------------------------------')

print('The MSE of NEW linear MultiLayerPerceptron is: ',MSE)

print('------------------------------------------')
````
